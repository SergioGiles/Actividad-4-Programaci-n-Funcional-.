```python
# Sustituir nulos con el promedio

def sustituir_nulos_promedio(df):
    
    # Seleccionar solo las columnas numéricas
    numeric_cols = df.select_dtypes(include=['number']).columns
    
    # Sustituir nulos con el promedio solo en columnas numéricas
    df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
    
    return df
```

```python
# Sustituir nulos con el promedio

df_promedio = sustituir_nulos_promedio(df)
print("\nDataFrame después de sustituir nulos con el promedio:")
print(df_promedio.head())
```

```python
# Sustituir nulos con la mediana

def sustituir_nulos_mediana(df):
    
    # Seleccionar solo las columnas numéricas
    numeric_cols = df.select_dtypes(include=['number']).columns
    
    # Sustituir nulos con la mediana solo en columnas numéricas
    df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
    
    return df
```

```python
# Sustituir nulos con la mediana
df_mediana = sustituir_nulos_mediana(df)
print("\nDataFrame después de sustituir nulos con la mediana:")
print(df_mediana.head())
```

```python
# Sustituir nulos con una constante
df_constante = sustituir_nulos_constante(df, 0)
print("\nDataFrame después de sustituir nulos con 0:")
print(df_constante.head())
```
